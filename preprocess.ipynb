```python
import numpy as np
import pandas as pd

data=pd.read_csv("all_drinks.csv")
print(data.columns)
```

```python
print(data.head())
```

```python
ingredient_cols = [col for col in data.columns if 'strIngredient' in col]

data['ingredients'] = data[ingredient_cols].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)

data['ingredients'] = data['ingredients'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
print(data['ingredients'])
```

```python
data['strDrink'] = data['strDrink'].str.lower()
data['ingredients'] = data['ingredients'].str.lower()

data['strCategory'] = data['strCategory'].str.lower()
data['strAlcoholic'] = data['strAlcoholic'].str.lower()
```

```python
data.count()
```

```python
df = data.drop_duplicates(subset='strDrink')
```

```python
df.count()
```

```python
df['isAlcoholic'] = df['strAlcoholic'].apply(lambda x: 1 if x == 'alcoholic' else 0)
```

```python
df.head()
```

```python
df = df[['strDrink', 'ingredients', 'strCategory', 'isAlcoholic']]
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(df['ingredients'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['strDrink']).drop_duplicates()

def recommend_cocktails(cocktail_name, cosine_sim=cosine_sim):
    idx = indices[cocktail_name]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    cocktail_indices = [i[0] for i in sim_scores]

    return df['strDrink'].iloc[cocktail_indices]

print(recommend_cocktails('margarita'))
```

```python
from collections import Counter

def recommend_for_favorites(favorites, num_recommendations=25, num_random=5):
    recommendations_count = Counter()

    # Pour chaque cocktail favori, obtenir des recommandations
    for favorite in favorites:
        similar_cocktails = recommend_cocktails(favorite)
        recommendations_count.update(similar_cocktails)

    top_recommendations = [cocktail for cocktail, _ in recommendations_count.most_common(num_recommendations)]

    random_cocktails = df['strDrink'].sample(num_random).tolist()
    final_recommendations = top_recommendations + random_cocktails

    return final_recommendations

#Simulation of utilisation
favorites = ['margarita', 'mojito', 'old fashioned']
print(recommend_for_favorites(favorites))
```
